```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

BASE_PATH = r"../Dashboard/data"


file_conv_hijos  = f"{BASE_PATH}/conversor_centiles_a_euros_hijos.csv"
file_conv_padres = f"{BASE_PATH}/conversor_centiles_a_euros_padres.csv"
file_tabla_ccaa  = f"{BASE_PATH}/ranking_ccaa_centil_padres_20.csv"
```

```python
read_opts = dict(sep=',', encoding='UTF-8')

conversor_hijos  = pd.read_csv(file_conv_hijos, **read_opts)
conversor_padres = pd.read_csv(file_conv_padres, **read_opts)
tabla_ccaa       = pd.read_csv(file_tabla_ccaa, **read_opts)
```

```python
datasets = {
    "conversor_hijos": conversor_hijos,
    "conversor_padres": conversor_padres,
    "tabla_ccaa": tabla_ccaa
}

for nombre, df in datasets.items():
    print(f"\n=== {nombre} ===")
    print("Shape:", df.shape)
    print("Columnas:", list(df.columns)[:12], '...')
```

```python
COL_PERC = "centil"
COL_EUROS = "renta"

tabla_conv_hijos = conversor_hijos[[COL_PERC, COL_EUROS]].dropna()

def centil_hijos_a_euros(c):
    fila = tabla_conv_hijos.loc[tabla_conv_hijos[COL_PERC] == c]
    if fila.empty:
        return np.nan
    return float(fila[COL_EUROS].iloc[0])

centil_hijos_a_euros(20), centil_hijos_a_euros(80)
```

```python
COL_PERC = "centil"
COL_EUROS = "renta"

tabla_conv_padres = conversor_padres[[COL_PERC, COL_EUROS]].dropna()

def centil_padres_a_euros(c):
    fila = tabla_conv_padres.loc[tabla_conv_padres[COL_PERC] == c]
    if fila.empty:
        return np.nan
    return float(fila[COL_EUROS].iloc[0])

centil_padres_a_euros(20), centil_padres_a_euros(80)
```

```python
COL_CCAA_NOMBRE = "ccaa"
COL_P_HIJOS_M = "centil_hijo_loess"

ranking_ccaa = (
    tabla_ccaa.groupby(COL_CCAA_NOMBRE)[COL_P_HIJOS_M]
    .mean()
    .reset_index()
    .sort_values(COL_P_HIJOS_M, ascending=False)
)

ranking_ccaa.head(10)
```
